```python
import pandas as pd
df = pd.DataFrame.from_csv('titanic_data.csv')
df.head()
```

```python
df['Pclass'] = df['Pclass'].map({3: 'Third', 2: 'Second', 1: 'First'})
df['Sex'] = df['Sex'].map({'male': 'Male', 'female': 'Female'})
df.head()
```

```python
def cutDF(df):
    return pd.cut(df,[0, 13, 20, 35, 55, 150], labels=['Child (0-12)', 'Teenager (13-19)', 'Young Adult (20-34)', \
                                                       'Adult (35-54)', 'Senior (55+)'])

df['Age Group'] = df[['Age']].apply(cutDF)
df.head(30)
```

```python
df.to_csv('new_titanic.csv')
```
