```python
import pandas as pd
df = pd.DataFrame.from_csv('titanic_data.csv')
df.head()
```

```python
df['Pclass'] = df['Pclass'].map({3: 'Third', 2: 'Second', 1: 'First'})
df.head()
```

```python
def cutDF(df):
    return pd.cut(df,[0, 13, 20, 35, 55, 150], labels=['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior'])

df['Category'] = df[['Age']].apply(cutDF)
df.head()
```

```python
df.to_csv('new_titanic.csv')
```
